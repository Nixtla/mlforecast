# DaskXGBForecast

> dask XGBoost forecaster

Wrapper of `xgboost.dask.DaskXGBRegressor` that adds a `model_` property that contains the fitted model and is sent to the workers in the forecasting step.

```python
#|export
import xgboost as xgb
```

```python
#|export
class DaskXGBForecast(xgb.dask.DaskXGBRegressor):
    @property
    def model_(self):
        model_str = self.get_booster().save_raw('ubj')
        local_model = xgb.XGBRegressor()
        local_model.load_model(model_str)
        return local_model
```
